```python
import pandas as pd
```

```python
jan_df = pd.read_csv('data/202201-bluebikes-tripdata.csv')
march_df = pd.read_csv('data/202203-bluebikes-tripdata.csv')
may_df = pd.read_csv('data/202205-bluebikes-tripdata.csv')
july_df = pd.read_csv('data/202207-bluebikes-tripdata.csv')
sept_df = pd.read_csv('data/202209-bluebikes-tripdata.csv')
nov_df = pd.read_csv('data/202211-bluebikes-tripdata.csv')
```

```python
may_df.head(5)
```

```python
def cleaning_df(df):
    '''Dropping null values, giving better description of column values, and tidying up column names'''

    df.drop_na(how='any',inplace=True)
    clean_df = df.rename(columns={'tripduration':'Trip Duration (s)', 'starttime':'Start Time', 'stoptime':'Stop Time',
                                  'start station id': 'Start Station ID','start station name':'Start Station Name','start station latitude':'Start Station Latitude',
                                  'start station longitude':'Start Station Longitude','end station id': 'End Station ID','end station name':'End Station Name',
                                  'end station latitude':'End Station Latitude','end station longitude':'End Station Longitude',
                                  'bikeid':'Bike ID','usertype':'User Type','postal code':'Postal Code'})
    

    return clean_df
```

```python
dfs = [jan_df,march_df,may_df,july_df,sept_df,nov_df]
strings = ['jan_df','march_df','may_df','july_df','sept_df','nov_df']
```

```python
for i, df in enumerate(dfs):
    name = strings[i]
    df = cleaning_df(df)
    df.to_csv(f'data/cleaned/{name}.csv')
```
